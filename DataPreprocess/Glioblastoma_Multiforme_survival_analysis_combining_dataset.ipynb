**Combine Clinical Data and mrna Seq Data**


***Mrna Data***

```python
# define the path of the mrna seq data
MRNA_PATH = "/content/drive/Shareddrives/1:1_Krish_Nachnani/2024/Glioblastoma Multiforme -genome data/gbm_tcga/data_mrna_seq_v2_rsem_zscores_ref_all_samples.txt"
```

```python
import pandas as pd
import numpy as np

# read the csv
df2 = pd.read_csv(MRNA_PATH, sep='\t')
df2.head()
```

```python
# Transpose df2 so that patient IDs become the index
df2_transposed = df2.set_index('Hugo_Symbol').transpose()
# Adjust the patient IDs in df2_transposed by removing the '-01' suffix
df2_transposed.index = df2_transposed.index.str.replace('-01', '')
df2_transposed.head()
```

***Clinical Data***

```python
# define the clinical data path
CLINICAL_PATH = "/content/drive/Shareddrives/1:1_Krish_Nachnani/2024/Glioblastoma Multiforme -genome data/gbm_tcga/data_clinical_patient.txt"
```

```python
# read the data
df1 = pd.read_csv(CLINICAL_PATH, sep='\t')
df1.head()
```

```python
print(df1.shape[0])
```

```python
# Create a dictionary from df1 mapping patient ID to 'Disease Free Status'
disease_free_status_dict = df1.set_index('Patient Identifier')['Disease Free Status'].to_dict()
disease_free_months_dict = df1.set_index('Patient Identifier')['Disease Free (Months)'].to_dict()
disease_free_age = df1.set_index('Patient Identifier')['Diagnosis Age'].to_dict()
```

***Map Indexes***

```python
# Map the 'Disease Free Status' to the transposed df2 based on patient IDs
df2_transposed['Disease Free Status'] = df2_transposed.index.map(disease_free_status_dict)
df2_transposed['Disease Free (Months)'] = df2_transposed.index.map(disease_free_months_dict)
df2_transposed['Diagnosis Age'] = df2_transposed.index.map(disease_free_age)
df2_transposed.head()
```

```python
df2_transposed.shape[0]
```

```python
# drop nas in Disease Free Status column
df_cleaned = df2_transposed.dropna(subset=['Disease Free Status'])
print(df_cleaned['Disease Free Status'].unique())
```

```python
# Drop rows with 'Not Available' in column 'A'
df_cleaned = df_cleaned[df_cleaned['Disease Free Status'] != '[Not Available]']
print(df_cleaned['Disease Free Status'].unique())
```

Encode disease free column

```python
# Select relevant columns and remove rows where DFS_STATUS is '[Not Available]'
sample_data = df_cleaned.copy()

# Encode DFS_STATUS to binary
sample_data.loc[sample_data['Disease Free Status'] == '0:DiseaseFree', 'DFS_STATUS_ENCODED'] = 0
sample_data.loc[sample_data['Disease Free Status'] == '1:Recurred/Progressed', 'DFS_STATUS_ENCODED'] = 1


# Convert DFS_MONTHS to float and handle '[Not Available]'
sample_data = sample_data[sample_data['Disease Free (Months)'] != '[Not Available]'].copy()
sample_data['Disease Free (Months)'] = sample_data['Disease Free (Months)'].astype(float)


# finally drop the Disease Free Status column
sample_data = sample_data.drop(['Disease Free Status'], axis=1)
sample_data.head()
```

```python
# Calculate the threshold for columns to be dropped
threshold = len(sample_data) * 0.5
# Drop columns with more than 50% NaNs
new_sample_data = sample_data.dropna(thresh=threshold, axis=1)
new_sample_data.isna().sum().sum()
```

```python
print(new_sample_data.shape[0])
print(new_sample_data.shape[1])
```

```python
new_sample_data.head()
```

```python
# Drop rows with any NaNs
new_sample_data = new_sample_data.dropna()
```
