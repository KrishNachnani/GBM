```python
import pandas as pd
import numpy as np
# read the data
data = pd.read_csv(PATH)
data.head()
```

```python
from sklearn.feature_selection import SelectKBest, f_classif
# define the components
#COMPONENTS = 50
#COMPONENTS = 20
COMPONENTS = 40

# select x and y values
X = data.drop(columns=['Disease Free (Months)', 'DFS_STATUS_ENCODED'])
y = data['DFS_STATUS_ENCODED']

# Select the top 50/20 features
selector = SelectKBest(f_classif, k=COMPONENTS)
X.columns = X.columns.astype(str)
X_new = selector.fit_transform(X, y)

# filter the dataframe with the selected features
selected_features = list(X.columns[selector.get_support()]) + ['Disease Free (Months)', 'DFS_STATUS_ENCODED']
print(selected_features)

# select training and testing data based on the filter
selected_data = data[selected_features]
```

```python
import os

FOLDER_NAME = "/content/drive/Shareddrives/1:1_Krish_Nachnani/2024/Glioblastoma Multiforme -genome data/40feature_select_csv"
FILE_NAME = "Select_K_Best_features.csv"
selected_data.to_csv(os.path.join(FOLDER_NAME, FILE_NAME), index=False)
```
